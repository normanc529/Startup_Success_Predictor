```python
# Initial imports
import os
import requests
import pandas as pd
from dotenv import load_dotenv
```

```python
# Load .env environment variables
load_dotenv()
```

```python
# Set Crunchbase API key and secret
api_key = os.getenv("NEW_RAPID_API_KEY")
```

```python
url = "https://crunchbase-crunchbase-v1.p.rapidapi.com/autocompletes"

querystring = {"query":""}

headers = {
    'x-rapidapi-key': api_key,
    'x-rapidapi-host': "crunchbase-crunchbase-v1.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```

```python
# Convert JSON string to a dictionary
crunchbase_data = response.json()
crunchbase_data
```

```python
# Create Dataframe 
cb_df = pd.DataFrame(crunchbase_data['entities'])
cb_df.tail()
```
